```python
import pandas as pd
```

Fonte de dados:

```python
dados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true', compression='zip')
```

O banco de dados possui 23814 observações e 877 variáveis:

```python
dados.shape
```

- **id** é uma variável *alfanumérica* e identifica uma única observação

- **tratamento** é uma variável *categórica nominal* com valores *'com_droga' e 'com_controle'*. Identifica as observações que fazem parte do grupo de estudo ou de controle

```python
dados['tratamento'].unique()
```

- **tempo** é uma variável *categórica ordinal* com valores *24, 72 e 48*. Representa o intervalo em uma unidade de tempo em que a droga foi administrada e o resultado medido

```python
dados['tempo'].unique()
```

- **dose** é uma variável *categórica nominal* com valores *'D1' e 'D2'*. Representa o esquema de dosagem de droga aplicada

```python
dados['dose'].unique()
```

- **droga** é uma variável *alfanumérica* e identifica a droga aplicada. Foram testadas 3289 drogas diferentes

```python
len(dados['droga'].unique())
```

- **g-0** .. **g-771** e **c-0** .. **c-99** são variáveis *numéricas contínuas*, variando de *-10* até *10*. Representam a expressão genética em relação à droga aplicada

```python
dados.filter(regex='^g')
```

```python
dados.filter(regex='^c')
```

```python
dados[dados.columns[5:]].min().min()
```

```python
dados[dados.columns[5:]].max().max()
```

Existem muito mais observações do grupo de estudo que do grupo de controle

```python
dados['tratamento'].value_counts(normalize=True).plot.bar()
```

A variável **tempo** possui valores mais parecidos, com um pouco mais de casos com o tempo de *48*


```python
dados['tempo'].value_counts(normalize=True).plot.bar()
```

O esquema de **dose** possui distribuição de valores praticamente equilibrada

```python
dados['dose'].value_counts(normalize=True).plot.bar()
```
